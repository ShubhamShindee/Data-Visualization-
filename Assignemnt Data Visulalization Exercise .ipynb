### Q.3 Read motorcycle and classic car productline sales data and show it using a multiline plot and take first 100 rows from dataframe .
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
new=df.head(100)
Motorcycle=new[new["PRODUCTLINE"]=="Motorcycles"]
Classic_cars=new[new["PRODUCTLINE"]=="Classic Cars"]
x1=Motorcycle["MONTH_ID"]
y1=Motorcycle["SALES"]
x2=Classic_cars["MONTH_ID"]
y2=Classic_cars["SALES"]
plt.figure(figsize=(10,5))
plt.plot(x1,y1,marker="o",color="Blue",linestyle="-",label="Motorcycles Sales Data")
plt.plot(x2,y2,marker="*",color="red",linestyle="-",label="Classic Cars Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("company Sales Data")
plt.legend(loc="best")
```

### Q.9 Read motorcycle and classic car of all months and display it using the Subplot and take first 100 rows from data frame.

```python
x1=Motorcycle["MONTH_ID"]
y1=Motorcycle["SALES"]
x2=Classic_cars["MONTH_ID"]
y2=Classic_cars["SALES"]

# for motorcycle
plt.figure(figsize=(12,8))
plt.subplot(3,1,1) ## no. of row , no. of column , position number
plt.plot(x1,y1,color="r",label="Motorcycle sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Data of Motorcycle")
plt.legend()

# for classic cars
plt.figure(figsize=(12,8))
plt.subplot(3,1,2)  ## no. of row , no. of column , position number
plt.plot(x2,y2,color="g",label="Classic cars sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Data of Classic cars")
plt.legend()

#for barplot of motocycle
plt.figure(figsize=(12,8))
plt.subplot(3,1,3)  ## no. of row , no. of column , position number
plt.bar(x1,y1,label="Motorcycle sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Barplot of Motorcycle")
plt.legend()
```

### Q.10 Calculate total sale data for motor cycle ,classic car, truck and bus,vintage car product and show it using a Pie chart take first 1000 rows from data-frame.

```python
rows=df.head(1000)
new=rows.groupby("PRODUCTLINE")
Motorcycle=new.get_group("Motorcycles")
Classic_cars=new.get_group("Classic Cars")
Vintage_cars=new.get_group("Vintage Cars")
truks=new.get_group("Trucks and Buses")
x3=Motorcycle["SALES"].sum()
x4=Classic_cars["SALES"].sum()
x5=Vintage_cars["SALES"].sum()
x6=truks["SALES"].sum()

plt.figure(figsize=(15,8))
plt.pie([x3,x4,x5,x6],labels=["Motorcycle","Classic_cars","Vintage_cars","Trucks and Buses"],autopct="%1.2f%%")
plt.legend(loc="lower right")
plt.show()
```
