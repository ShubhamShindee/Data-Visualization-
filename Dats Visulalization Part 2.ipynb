```python
df=sns.load_dataset("iris")
df.head()
```

```python
df.columns
```

```python
help(sns)
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df,marker="o")
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species",style="species",markers=True)
```

```python
help(sns.lineplot)
```

### scatterlot

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=df)
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=df,hue="species")
```

### barplot

```python
sns.barplot(x="sepal_length",y="sepal_width",data=df)
```

#### Histogram

```python
sns.histplot(x="sepal_length",data=df)
```

```python
sns.displot(x="sepal_length",data=df)
```

```python
sns.boxplot(x="sepal_width",data=df)
```

```python
sns.jointplot(x="sepal_width",y="sepal_length",data=df)
```

```python
sns.jointplot(x="sepal_width",y="sepal_length",data=df,kind="hex")
```

```python
sns.regplot(x="sepal_width",y="sepal_length",data=df)
```

```python
sns.swarmplot(x="species",y="sepal_width",data=df)
```

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(x_vars=['sepal_length', 'sepal_width', 'petal_length'],y_vars="petal_width",data=df)
```
